## Creazione Nuovo Dataset per Binary Classification

```python
class_0 = iris_df[iris_df['class']== 0] #Restituisce le righe con classe 0
class_1 = iris_df[iris_df['class']== 1] #Restituisce le righe con classe 1
class_2 = iris_df[iris_df['class']== 2] #Restituisce le righe con classe 2

class_0.head()#check
```

### Primo Classificatore
- Classe 0 --> Classe Positiva
- (Classe1 + Classe2)-->Classe Negativa

```python
classe_negativa_1 = np.r_[class_1,class_2] #concateniamo sulle righe
classe_negativa_1[:,-1] = 0 #sovrascrivo l'etichetta da 1 e 2 a 0
classe_positiva_1 = class_0.values
classe_positiva_1[:,-1] = 1 #sovrascrivo l'etichetta da 0 a 1

dataset1 = np.r_[classe_negativa_1 ,classe_positiva_1 ] #concateniamo sulle righe
X_1 = np.c_[np.ones((dataset1.shape[0],1)),dataset1[:,:4]]#prendiamo le prime 4 colonne del dataset1
y_1 = dataset1[:,-1].reshape(-1,1)#prendiamo l'ultima colonna del dataset1

print(X_1)
print(y_1)
```

### Secondo Classificatore
- Classe 1 --> Classe Positiva
- (Classe0 + Classe2)-->Classe Negativa

```python
classe_negativa_2 = np.r_[class_0,class_2] #concateniamo sulle righe
classe_negativa_2[:,-1] = 0 #sovrascrivo l'etichetta da 0 e 2 a 0
classe_positiva_2 = class_1.values

dataset2 = np.r_[classe_negativa_2 ,classe_positiva_2 ] #concateniamo sulle righe
X_2 = np.c_[np.ones((dataset2.shape[0],1)),dataset2[:,:4]]#prendiamo le prime 4 colonne del dataset1
y_2 = dataset2[:,-1].reshape(-1,1)#prendiamo l'ultima colonna del dataset1

print(X_2)
print(y_2)
```

### Terzo Classificatore
- Classe 2 --> Classe Positiva
- (Classe0 + Classe1)-->Classe Negativa

```python
classe_negativa_3 = np.r_[class_0,class_1] #concateniamo sulle righe
classe_negativa_3[:,-1] = 0 #sovrascrivo l'etichetta da 0 e 1 a 0
classe_positiva_3 = class_2.values
classe_positiva_3[:,-1] = 1 #sovrascrivo l'etichetta da 2 a 1

dataset3 = np.r_[classe_negativa_3 ,classe_positiva_3 ] #concateniamo sulle righe
X_3 = np.c_[np.ones((dataset3.shape[0],1)),dataset3[:,:4]]#prendiamo le prime 4 colonne del dataset1
y_3 = dataset3[:,-1].reshape(-1,1)#prendiamo l'ultima colonna del dataset1

print(X_3)
print(y_3)
```

## Addestramento Classificatori

```python
def sigmoid(z):
    return(1 / (1 + np.exp(-z)))
```

```python
def costFunction(theta, X, y):
    m = y.size
    h = sigmoid(X.dot(theta))
    
    J = -1*(1/m)*(np.log(h).T.dot(y) + np.log(1-h).T.dot(1-y))
               
    if np.isnan(J[0]):
        return(np.inf)
    return(J[0])
```

```python
def gradient(theta, X, y):
    m = y.size
    h = sigmoid(X.dot(theta.reshape(-1,1)))
    
    grad =(1/m)*X.T.dot(h-y)

    return(grad.flatten())
```

```python
initial_theta_1 = np.zeros((X_1.shape[1],1))#vettore parametri 5x1
initial_theta_2 = np.zeros((X_2.shape[1],1))#vettore parametri 5x1
initial_theta_3 = np.zeros((X_3.shape[1],1))#vettore parametri 5x1

res_1 = minimize(costFunction, initial_theta_1, args = (X_1, y_1), method = 'SLSQP', jac = gradient, options = {'maxiter' : 1000})
res_2 = minimize(costFunction, initial_theta_2, args = (X_2, y_2), method = 'SLSQP', jac = gradient, options = {'maxiter' : 1000})
res_3 = minimize(costFunction, initial_theta_3, args = (X_3, y_3), method = 'SLSQP', jac = gradient, options = {'maxiter' : 1000})
```

```python
theta_1 = res_1.x.reshape(-1,1) #Parametri dei 3 classificatori
theta_2 = res_2.x.reshape(-1,1)
theta_3 = res_3.x.reshape(-1,1)
```
