# COVID-19 Analysis - The Italian case 

```python
import numpy as np

import matplotlib.pyplot as plt 
import operator
from matplotlib.pyplot import figure

import pandas as pd  
import seaborn as sns 

%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

import time
```

```python
path = './COVID-19/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv'

dataframe_andamento_nazionale = pd.read_csv(path, delimiter = ',')
```

```python
dataframe_andamento_nazionale = dataframe_andamento_nazionale.drop(columns = ['casi_testati', 'note_it', 'note_en'])
```

## Plot the 'Pearson' Correlation Matrix

$$\large \rho_{X,Y} = \frac{\sum_{i} (x_{i} - \mu_{X})(y_{i}-\mu_{Y})}{\sqrt{\sum_{j}(x_{j}-\mu_{X})^2 \sum_{k} (y_{k}-\mu_{Y})^2}}$$

$$\large -1 \le \rho_{X,Y} \le 1$$

```python
f, (ax1) = plt.subplots(1, 1, figsize=(24,20))
correlation_matrix = dataframe_andamento_nazionale.corr(method='pearson')
sns.heatmap(correlation_matrix, cmap='coolwarm_r', annot = True, annot_kws={'size':25}, ax=ax1)
ax1.set_title("COVID-19 Andamento Nazionale - Correlation Matrix (Pearson Index)", fontsize=18)
plt.show()
```

## Plot the 'Spearman' Correlation Matrix

```python
f, (ax1) = plt.subplots(1, 1, figsize=(24,20))
correlation_matrix = dataframe_andamento_nazionale.corr(method='spearman')
sns.heatmap(correlation_matrix, cmap='coolwarm_r', annot = True, annot_kws={'size':25}, ax=ax1)
ax1.set_title("COVID-19 Andamento Nazionale - Correlation Matrix (Spearman Index)", fontsize=18)
plt.show()
```

## Legend

* **ricoverati_con_sintomi** : Ricoverati con sintomi / Hospitalised patients with symptoms
* **terapia_intensiva** : Ricoverati in terapia intensiva / Intensive Care
* **totale_ospedalizzati** : Totale ospedalizzati (ricoverati_con_sintomi + terapia_intensiva) / Total hospitalised patients
* **isolamento_domiciliare** : Persone in isolamento domiciliare / Home confinement
* **totale_positivi** : Totale attualmente positivi (ospedalizzati + isolamento domiciliare) / Total amount of current positive cases (Hospitalised patients + Home confinement)
* **variazione_totale_positivi** : Variazione del totale positivi (totale_positivi giorno corrente - totale_positivi giorno precedente)	/ News amount of current positive cases (totale_positivi current day - totale_positivi previous day)
* **nuovi_positivi** : Nuovi attualmente positivi (totale_casi giorno corrente - totale_casi giorno precedente)	/ News amount of current positive cases (totale_casi current day - totale_casi previous day)
* **dimessi_guariti** : Persone dimesse guarite	(dimessi dall'ospedale + guariti a casa) / Recovered
* **deceduti**	Persone decedute / Death
* **totale_casi** : Totale casi positivi / Total amount of positive cases
* **tamponi** : Totale tamponi / Tests performed

## Dates

```python
import datetime

start_date = datetime.date(2020, 2, 24)

date = []
for i in range(dataframe_andamento_nazionale.shape[0]):
    date.append(start_date + datetime.timedelta(days = i))
```

```python
date[15:]
```

# 1) Modello totale positivi giorno precedente (lag = 1) e nuovi positivi (E. Cimino)

```python
X_tp_1 = dataframe_andamento_nazionale['totale_positivi']
y_np_1 = dataframe_andamento_nazionale['nuovi_positivi']
```

```python
fig = plt.figure(figsize=(18,10))

ax1 = fig.add_subplot(211)
ax1.scatter(date, X_tp_1, label='positive cases', s=30, c='c', marker='o')
ax1.set_title('Total amount of current positive cases', fontsize = 14)
ax1.set_xlabel('days', fontsize = 14)
ax1.axvspan(date[15], date[-1], color='red', alpha=0.2, label =  "Lockdown")
#ax1.set_yscale('log')
#ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
ax1.minorticks_on()
ax1.legend()

ax2 = fig.add_subplot(212)
ax2.scatter(date, y_np_1, label='new positive cases', s=30, c='r', marker='o')
ax2.set_title('New amount of current positive cases', fontsize = 14)
ax2.set_xlabel('days', fontsize = 14)
ax2.axvspan(date[15], date[15+11], color='red', alpha=0.2, label =  "Lockdown")
ax2.axvspan(date[15+11], date[-1], color='green', alpha=0.2, label =  "Lockdown - after 11 days")
#ax1.set_yscale('log')
#ax2.xaxis.set_major_locator(MaxNLocator(integer=True))
ax2.minorticks_on()
ax2.legend(loc = 'upper left')

fig.tight_layout()

plt.show()
```

```python
f, (ax1) = plt.subplots(1, 1, figsize=(15,12))
correlation_matrix = pd.DataFrame(np.c_[X_tp_1, X_tp_1.shift(periods = 1), y_np_1], columns = ['totale positivi', 'totale positivi (t-1)', 'nuovi positivi']).corr(method='pearson')
sns.heatmap(correlation_matrix, cmap='coolwarm_r', annot = True, annot_kws={'size':25}, ax=ax1)
ax1.set_title("Use case 1 - Correlation Matrix (Pearson Index)", fontsize=18)
plt.show()
```
